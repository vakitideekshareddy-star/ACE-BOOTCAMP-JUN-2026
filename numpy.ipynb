Arithmetic (Operators and Functions)
Scalar Arthmetics (Operations perform using a single value

```python
import numpy as np
arr = np.array([[1,4,5,8,9],[1,64,25,18,9],[18,4,56,8,29]])
arr1 = np.array([2,8,10,16,18])

print("Multiplication:",arr*2)
print("Addition:",arr+5)
print(np.add(2,arr1))#Scalar Arthmetic Function
print("Subtraction:",arr-1)
print("Division:",arr/2)
print(np.divide(2,arr1))#Scalar Arthmetic Function
print("Modulus:",arr%2)
print("Power:",arr**2)
print(arr//2)


#Vector Arthmetic
print(np.sum(arr))
print(np.sum(arr1))


print("Actual Array:\n",arr)
print("Sum of the Elements using vector:\n",np.sum(arr,0))
```

```python
import numpy as np

arr=np.array([
    [1,2],
    [3,4]
])
arr1=np.array([
    [1,2],
    [3,4]
])
arr1.flags.writeable = False #read only
print(arr1)
#arr1[0,0]=100
print("Multiplication:\n",arr @ arr1)
print("Multiplication:\n",np.matmul(arr,arr1))
print("Addition:\n",arr+arr1)
print("Subtraction:\n",arr-arr1)
print("Division:\n",arr/arr1)
print("Modulus:\n",arr%arr1)
print("Power:\n",arr**arr1)
print(np.mean(arr))
print(np.std(arr))
print(np.var(arr))
print(np.min(arr))
print(np.max(arr))
print(np.argmin(arr))
print(np.argmax(arr))
```
